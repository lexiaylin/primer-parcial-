```python
class RGBImageFolder(datasets.ImageFolder):
    def __getitem__(self, index):
        path, label = self.samples[index]
        with open(path, 'rb') as f:
            img = Image.open(f).convert('RGB')  # <- FORZAMOS RGB
        if self.transform is not None:
            img = self.transform(img)
        return img, label

dataset = RGBImageFolder(root='/content/dataset', transform=transform)

# Definir proporciones (80% entrenamiento, 20% validación)
train_ratio = 0.8
train_size = int(train_ratio * len(dataset))
val_size = len(dataset) - train_size

# Dividir el dataset de forma aleatoria pero estratificada
# Para mantener proporción similar de clases en ambos conjuntos
train_dataset, val_dataset = random_split(
    dataset,
    [train_size, val_size],
    generator=torch.Generator().manual_seed(42)  # Semilla para reproducibilidad
)

# Verificar la división
print("Clases encontradas:", dataset.classes)  # Mostrará ['fake', 'real']
print("Cantidad total de imágenes:", len(dataset))
print("Tamaño entrenamiento:", len(train_dataset))
print("Tamaño validación:", len(val_dataset))
```

```python
# 1. Cargar modelo preentrenado
modelo = models.resnet18(weights=models.ResNet18_Weights.IMAGENET1K_V1)  # Forma actualizada de cargar pesos preentrenados
modelo.to(device)
# 2. Congelar capas convolucionales (no se entrenarán)
for param in modelo.parameters():
    param.requires_grad = False

# 3. Modificar la capa final completamente conectada
num_features = modelo.fc.in_features
modelo.fc = nn.Sequential(
    nn.Linear(num_features, 256),  # Capa intermedia adicional
    nn.ReLU(),                     # Función de activación
    nn.Dropout(0.5),               # Dropout para regularización
    nn.Linear(256, 2)              # Capa final (2 clases: real vs fake)
)

# 4. Mostrar arquitectura modificada
print("Arquitectura ResNet18 modificada:")
print(modelo)

# 5. Verificar parámetros entrenables
print("\nParámetros que se entrenarán:")
for name, param in modelo.named_parameters():
    if param.requires_grad:
        print(name)
```

```python
# Definir función de pérdida y optimizador
criterio = nn.CrossEntropyLoss()
optimizador = torch.optim.Adam(modelo.parameters(), lr=0.001)  # Entrenar TODAS las capas entrenables

# Número de épocas (puedes empezar con 10 e incrementar si es necesario)
EPOCAS = 10
```

```python
from torch.utils.data import DataLoader, random_split

# Suponiendo que ya tienes `dataset` desde datasets.ImageFolder
train_ratio = 0.8
train_size = int(train_ratio * len(dataset))
val_size = len(dataset) - train_size

train_dataset, val_dataset = random_split(dataset, [train_size, val_size])

batch_size = 32

train_loader = DataLoader(
    train_dataset,
    batch_size=batch_size,
    shuffle=True,
    num_workers=2,
    pin_memory=True
)

val_loader = DataLoader(
    val_dataset,
    batch_size=batch_size,
    shuffle=False,
    num_workers=2,
    pin_memory=True
)
```

```python
# Historiales para métricas
train_loss_hist, val_loss_hist = [], []
train_acc_hist, val_acc_hist = [], []

for epoca in range(EPOCAS):
    modelo.train()
    running_loss, correct, total = 0.0, 0, 0

    for imagenes, etiquetas in train_loader:
        imagenes, etiquetas = imagenes.to(device), etiquetas.to(device)

        optimizador.zero_grad()
        outputs = modelo(imagenes)
        loss = criterio(outputs, etiquetas)
        loss.backward()
        optimizador.step()

        running_loss += loss.item()
        _, predicted = torch.max(outputs.data, 1)
        total += etiquetas.size(0)
        correct += (predicted == etiquetas).sum().item()

    train_loss = running_loss / len(train_loader)
    train_acc = 100 * correct / total

    modelo.eval()
    val_loss_total, correct_val, total_val = 0.0, 0, 0
    with torch.no_grad():
        for imagenes, etiquetas in val_loader:
            imagenes, etiquetas = imagenes.to(device), etiquetas.to(device)
            outputs = modelo(imagenes)
            loss = criterio(outputs, etiquetas)
            val_loss_total += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            total_val += etiquetas.size(0)
            correct_val += (predicted == etiquetas).sum().item()

    val_loss = val_loss_total / len(val_loader)
    val_acc = 100 * correct_val / total_val

    train_loss_hist.append(train_loss)
    val_loss_hist.append(val_loss)
    train_acc_hist.append(train_acc)
    val_acc_hist.append(val_acc)

    print(f"Época {epoca+1}/{EPOCAS} - "
          f"Train Loss: {train_loss:.4f} | Train Acc: {train_acc:.2f}% | "
          f"Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.2f}%")
```

```python
# Guardar el modelo completo (arquitectura + pesos)
torch.save(modelo.state_dict(), "modelo_real_vs_fake.pth")  # Recomendado: guardar solo los pesos

# Opcional: Guardar también el optimizador para continuar entrenamiento después
checkpoint = {
    'model_state_dict': modelo.state_dict(),
    'optimizer_state_dict': optimizador.state_dict(),
    'epoch': EPOCAS,
    'train_acc_hist': train_acc_hist,
    'val_acc_hist': val_acc_hist
}
torch.save(checkpoint, "checkpoint_real_vs_fake.pth")

# Descargar los archivos desde Google Colab
from google.colab import files
files.download("modelo_real_vs_fake.pth")
files.download("checkpoint_real_vs_fake.pth")  # Opcional

print("\n¡Entrenamiento completado con éxito! ✅")
print(f"Modelo guardado como: modelo_real_vs_fake.pth")
print(f"Precisión final - Entrenamiento: {train_acc_hist[-1]:.2%} | Validación: {val_acc_hist[-1]:.2%}")
```

```python
def evaluar_modelo(model, dataloader, criterion):
    model.eval()
    correctos = 0
    total = 0
    loss_total = 0
    all_preds = []
    all_labels = []

    with torch.no_grad():
        for images, labels in dataloader:
            images = images.to(device)
            labels = labels.to(device)

            outputs = model(images)
            loss = criterion(outputs, labels)
            loss_total += loss.item()

            _, predicciones = torch.max(outputs.data, 1)
            total += labels.size(0)
            correctos += (predicciones == labels).sum().item()

            # Guardar para métricas adicionales
            all_preds.extend(predicciones.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    # Cálculo de métricas
    precision = 100 * correctos / total
    perdida_promedio = loss_total / len(dataloader)

    # Métricas adicionales
    from sklearn.metrics import classification_report, confusion_matrix
    print("\n" + "="*60)
    print("📊 EVALUACIÓN COMPLETA DEL MODELO")
    print("="*60)
    print(f"\n🔹 Precisión global: {precision:.2f}%")
    print(f"🔹 Pérdida promedio: {perdida_promedio:.4f}\n")

    # Reporte de clasificación
    print("📝 Reporte de clasificación:")
    print(classification_report(all_labels, all_preds, target_names=['Real', 'IA']))

    # Matriz de confusión
    conf_matrix = confusion_matrix(all_labels, all_preds)
    print("🧮 Matriz de confusión:")
    print(conf_matrix)

    # Visualización de matriz de confusión
    plt.figure(figsize=(8, 6))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Real', 'IA'],
                yticklabels=['Real', 'IA'])
    plt.title('Matriz de Confusión - Real vs IA')
    plt.ylabel('Verdaderos')
    plt.xlabel('Predicciones')
    plt.show()

    return {
        'precision': precision,
        'perdida': perdida_promedio,
        'reporte': classification_report(all_labels, all_preds, output_dict=True),
        'matriz_confusion': conf_matrix
    }
```
